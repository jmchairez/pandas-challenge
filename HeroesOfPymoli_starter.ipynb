```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
data_file = os.path.join("..", "Resources", "purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
df_purc_data = pd.read_csv(data_file)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_count = len(df_purc_data["Item Name"].unique())
```

```python
average = round(df_purc_data["Price"].mean(), 2)
```

```python
tot_purc = df_purc_data["Purchase ID"].count()
```

```python
tot_rev = df_purc_data["Price"].sum()
```

```python
summary_data = pd.DataFrame(
    {"Number of Unique Items": [unique_count], 
     "Average Price": "$"+str(average), 
     "Number of Purchases": [tot_purc], 
     "Total Revenue": [tot_rev]})

org_data = summary_data.loc[:, ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

org_data.loc[:,"Total Revenue"] = org_data["Total Revenue"].astype(float).map("${:,.2f}".format)
org_data
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_item = df_purc_data.groupby("Item ID")["Item Name"]
pop_item = pop_item.max()
```

```python
pop_count = df_purc_data["Item ID"].value_counts()
```

```python
pop_price = df_purc_data.groupby("Item ID")["Price"].mean()
```

```python
pop_total_value = df_purc_data.groupby("Item ID")["Price"].sum()
```

```python
pop_summary = pd.DataFrame({"Item Name": pop_item, 
                            "Purchase Count": pop_count, 
                            "Item Price": pop_price, 
                            "Total Purchase Value": pop_total_value}).sort_values(by="Purchase Count", ascending=False).head()

org_pop = pop_summary.loc[:, ["Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]]

org_pop.loc[:,"Item Price"] = org_pop["Item Price"].astype(float).map("${:,.2f}".format)
org_pop.loc[:,"Total Purchase Value"] = org_pop["Total Purchase Value"].astype(float).map("${:,.2f}".format)
org_pop
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
prof_summary = pd.DataFrame({"Item Name": pop_item, 
                            "Purchase Count": pop_count, 
                            "Item Price": pop_price, 
                            "Total Purchase Value": pop_total_value}).sort_values(by="Total Purchase Value", ascending=False).head()

org_prof = prof_summary.loc[:, ["Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]]

org_prof.loc[:,"Item Price"] = org_prof["Item Price"].astype(float).map("${:,.2f}".format)
org_prof.loc[:,"Total Purchase Value"] = org_prof["Total Purchase Value"].astype(float).map("${:,.2f}".format)
org_prof
```
